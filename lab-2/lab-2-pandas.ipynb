```python
def get_author(url):
    test_string = url
    return test_string.split('/')[-1].split('.')[1:][0]
```

```python
df['author'] = df.author_url.map(get_author)
```

```python
df.genre_urls.head()
```

```python
test_genre = df['genre_urls'][1]
genres = test_genre.split('|')
print(genres)
for i in range(len(genres)):
    genres[i] = genres[i].split('/')[-1]
genres
```

```python
def split_and_join_genres(url):
    test_genre = url
    genres = test_genre.split('|')
    for i in range(len(genres)):
        genres[i] = genres[i].split('/')[-1]
    s = ', '
    return s.join(genres)
```

```python
split_and_join_genres("/genres/young-adult|/genres/science-fiction")
```

```python
split_and_join_genres("")
```

```python
df['genres']=df.genre_urls.map(split_and_join_genres)
df.head()
```
